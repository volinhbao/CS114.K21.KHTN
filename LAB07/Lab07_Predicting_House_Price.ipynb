# Tạo một model đơn giản để dự đoán giá nhà từ diện tích căn nhà



Ở bước này ta sẽ tiến hành train một model regression đơn giản. Vì ta không muốn
model sẽ học trên dữ liệu dùng để test nên ta phải chia tập dữ liệu làm 2: training set và test set. Ta sẽ cho model học trên tập training, sau đó dùng kết quả thu được kiểm tra lại trên tập test set(model chưa được học).

Ở đây ta chia training set và test set theo tỉ lệ 80-20

**Lưu ý:** ta gán 1 con số bất kỳ cho ```seed``` để mỗi lần chạy ta đều thu được bộ training set và test set giống nhau

````python
training_set, test_set = sales.random_split(.8,seed=42)
````

## Train model (1 feature)

Trong phạm vi bài lab này ta sẽ tạo một model đơn giản dự đoán giá dựa trên diện tích (square feet), ta gọi model này là ```sqft_model```

Thuật toán ML sử dụng là Linear Regression(LR).

Để khởi tạo một model LR (hoặc bất kỳ), ta dùng lệnh ```turicreate.<tên thuật toán>.create()```

Ta cần truyền vào:
- tập dữ liệu dùng để train(```training_set```)
- target: xác định xem model này dùng để dự đoán cái gì(output) - ở đây là giá nhà(```price```)
- features: những feature sẽ sử dụng để train model
- Ở đây mặc định hàm sẽ tạo validation set (tỉ lệ 5% của training set), nếu muốn bỏ qua bước tạo validation set ta có thể set ```validation_set=None```

````python
sqft_model = turicreate.linear_regression.create(training_set,target='price',features=['sqft_living'])
````

## Đánh giá chất lượng của model

Để đánh giá model, ta xem giá trị RMSE của model trên tập test

````python
# Xem giá trị trung bình của feature 'price' trên tập test
print (test_set['price'].mean())
````

Ta dùng lệnh evaluate() để tính RMSE cho dữ liệu trên tập test

````python
print (sqft_model.evaluate(test_set))
````

# Trực quan hóa kết quả thu được

````python
sqft_model.coefficients
````

Ta import thư viện ```matplotlib``` để trực quan hóa model bằng đồ thị

````python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(test_set['sqft_living'],test_set['price'],'.',
        test_set['sqft_living'],sqft_model.predict(test_set),'-')
````

# Tạo một model dự đoán giá nhà dựa trên nhiều feature

Ta mở rộng không chỉ xây dựng model dự đoán giá nhà trên một feature là diện tích.

## Train model (6 features)

Ta cũng xây dựng một model LR, nhưng dùng 6 features đã nêu trên thay vì chỉ dùng diện tích.

````python
my_features_model = turicreate.linear_regression.create(training_set,target='price',features=my_features)
````

# So sánh model đơn giản(1 feature) với model phức tạp hơn (6 features)

Sau khi training, ta tiến hành so sánh kết quả thu được từ việc training model với 1 feature và 6 feature

````python
print (my_features)
````

````python
print (sqft_model.evaluate(test_set))
print (my_features_model.evaluate(test_set))
````

Có thể thấy kết quả thu được từ việc training 6 features tốt hơn 1 feature (error thấp hơn)
